비용함수(Cost Function) : 손실함수(Loss Function), 훈련하는 동안의 결과값과 실제값(정답)과의 차이를 측정

image.png

```python
w_range = np.arange(0.1, 4.1, 0.1)
costs = []
for w in w_range:
    h = w * X
    cost = 1 / 50 * np.sum((h-Y) ** 2) # 평균 제곱 오차(MSE, Mean Squred Error)
    costs.append(cost)
plt.plot(w_range, costs, 'r.')  # X축: w(기울기), Y축: 오차(costs)
plt.show()
```

```python
len(w_range), w_range[20]
```

```python
w1 = w_range[:20]
c1 = costs[:20]
w1, c1
```

```python
plt.plot(w_range, costs, 'r.') # X축: w(기울기), Y축: 오차(costs)

# 실선을 그림, 실선이 기울기
for w, cost in zip(w1, c1):
    h = w * X
    gradient = 2 / 50 * np.sum( (h-Y) * X )  # 평균 제곱 오차(MSE, Mean Squred Error)의 기울기
    plt.plot(w_range, gradient*(w_range-w)+cost)
    plt.axis([0, 4, 0, 3000])
plt.show()
```

```python
# 일부만 그림
for w, cost in zip(w_range[18:21], costs[18:21]):
    plt.plot(w_range, costs, 'r.')
    h = w * X
    gradient = 2 / 50 * np.sum( (h-Y) * X )
    plt.plot(w_range, gradient*(w_range-w)+cost)
    plt.title(f'w_range: {w:.1f}, costs: {cost:.3f}, gradient: {gradient:.3f}')
    plt.axis([0, 4, 0, 3000])
    plt.show()
```

최소제곱법(Ordinary Least Squares)으로 구한 방법과 비교

```python
# 최소제곱법(Ordinary Least Squares)으로 기울기(w)와 절편(b) 구함
def OLS(x, y):
    w = np.sum((x-x.mean()) * (y-y.mean())) / np.sum((x-x.mean()) ** 2)
    b = y.mean() - w * x.mean()
    return w, b
```

```python
w, b = OLS(X, Y)
w, b
```

```python
Y_pred = X * w + b
```

```python
#plt.figure(figsize=(10, 5))

#plt.subplot(1, 2, 1)
#plt.scatter(X, Y, 'b.', label='Sample' )
plt.plot(X, Y, 'b.', label='Sample' )
plt.plot(X, Y_pred, c='red', label=f'Y={w:.2f}x+{b:.2f}')
plt.legend()
plt.title('Linear Regression-Ordinary Least Squares')
```

# 인공신경망 - 다층 퍼셉트론

```python
import numpy as np

# 가중치와 바이어스
w11 = np.array([-2, -2]) # 입력층 - 은닉층 n1의 가중치
w12 = np.array([2, 2])   # 입력층 - 은닉층 n2의 가중치
w2 = np.array([1, 1])    # 은닉층(n1, n2) - 출력층의 가중치
b1 = 3                   # 은닉층 n1의 바이어스
b2 = -1                  # 은닉층 n2의 바이어스
b3 = -1                  # 출력층의 바이어스

# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1

# NAND 게이트
def NAND(x1,x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1,x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트
def AND(x1,x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트
def XOR(x1,x2):
    return AND(NAND(x1, x2),OR(x1,x2))


# x1, x2 값을 번갈아 대입해 가며 최종 값 출력
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```
